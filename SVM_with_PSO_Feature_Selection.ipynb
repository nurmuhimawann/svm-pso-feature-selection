```python
def manual_label_encoder(df):
    # define catogory_mapping
    categorical_columns = ['Level']

    # Create a mapping dictionary for each categorical column
    column_mappings = {
        'Level' : {'Low': 0, 'Medium': 1, 'High': 2,}}

    for col in categorical_columns:
        mapping = column_mappings.get(col, {})  # Get the mapping for the current column
        df[col] = df[col].map(mapping)

        return df

df = manual_label_encoder(data)
df.head()
```

```python
# assign values
X = df.drop(['Level'], axis = 1).values # feature
Y = df['Level'].values # label
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
import numpy as np
import pyswarms as ps
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Fungsi untuk evaluasi fitur
def evaluate_features(features, X_train, X_test, y_train, y_test):
    print(f'features {features}')
    clf = SVC(kernel='linear')
    clf.fit(X_train[:, features], y_train)
    y_pred = clf.predict(X_test[:, features])
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Fungsi untuk inisialisasi partikel
def initialize_particles(num_particles, num_features):
    particles = np.random.randint(2, size=(num_particles, num_features))
    return particles

# Fungsi untuk menghitung nilai fitness
def calculate_fitness(particles, X_train, X_test, y_train, y_test):
    fitness = []
    for particle in particles:
        features = np.where(particle == 1)[0]
        accuracy = evaluate_features(features, X_train, X_test, y_train, y_test)
        fitness.append(accuracy)
    return np.array(fitness)

# Fungsi untuk pembaruan posisi partikel
def update_particles(particles, velocities, personal_best, global_best, w, c1, c2):
    r1 = np.random.rand(particles.shape[0], particles.shape[1])
    r2 = np.random.rand(particles.shape[0], particles.shape[1])

    velocities = w * velocities + c1 * r1 * (personal_best - particles) + c2 * r2 * (global_best - particles)
    particles = particles + velocities

    return particles, velocities

# Fungsi PSO untuk seleksi fitur
def pso_feature_selection(X_train, X_test, y_train, y_test, num_particles=30, num_iterations=2, w=0.5, c1=1.5, c2=1.5):
    num_features = X_train.shape[1]
    particles = initialize_particles(num_particles, num_features)
    velocities = np.zeros_like(particles)
    personal_best = particles.copy()
    fitness = calculate_fitness(particles, X_train, X_test, y_train, y_test)
    personal_best_fitness = fitness.copy()
    global_best_index = np.argmax(fitness)
    global_best = particles[global_best_index].copy()

    for iteration in range(num_iterations):
        particles, velocities = update_particles(particles, velocities, personal_best, global_best, w, c1, c2)
        fitness = calculate_fitness(particles, X_train, X_test, y_train, y_test)

        # Pembaruan personal best
        update_personal_best = fitness > personal_best_fitness
        personal_best[update_personal_best] = particles[update_personal_best]
        personal_best_fitness[update_personal_best] = fitness[update_personal_best]

        # Pembaruan global best
        global_best_index = np.argmax(fitness)
        if fitness[global_best_index] > calculate_fitness(global_best.reshape(1, -1), X_train, X_test, y_train, y_test):
            global_best = particles[global_best_index].copy()

        print(f"Iteration {iteration + 1}/{num_iterations} - Best Accuracy: {fitness[global_best_index]}")

    selected_features = np.where(global_best == 1)[0]
    return selected_features

# Contoh penggunaan PSO untuk seleksi fitur
selected_features = pso_feature_selection(X_train, X_test, y_train, y_test)
print("Selected Features:", selected_features)

# mwlatih model
clf = SVC(kernel='linear')
clf.fit(X_train[:, selected_features], y_train)

# Menguji model pada data uji
y_pred = clf.predict(X_test[:, selected_features])

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy}")

# Contoh penggunaan PSO untuk seleksi fitur
# (Pastikan untuk mengganti dataset dan target dengan data yang sesuai)
# X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
# selected_features = pso_feature_selection(X_train, X_test, y_train, y_test)
# print("Selected Features:", selected_features)
```
